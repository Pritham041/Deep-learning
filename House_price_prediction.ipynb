## Data exploration

### Subtask:
Explore the dataset to understand its structure, identify missing values, and analyze the distribution of variables.


**Reasoning**:
I need to explore the dataset, identify missing values, analyze the distribution of variables, and perform the other requested analyses.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data Types
print(df.dtypes)

# Missing Values
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
print(missing_percentage[missing_percentage > 0])

plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()

# Descriptive Statistics
numerical_features = df.select_dtypes(include=['int64', 'float64'])
print(numerical_features.describe())
for col in numerical_features.columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')
    plt.show()


# Categorical Feature Analysis
categorical_features = df.select_dtypes(include=['object'])
for col in categorical_features.columns:
    print(f"Unique values for {col}: {df[col].unique()}")
    print(df[col].value_counts())
    plt.figure(figsize=(10, 5))
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45, ha='right')
    plt.show()

# Target Variable Analysis
plt.figure(figsize=(8, 4))
sns.histplot(df['SalePrice'], kde=True)
plt.title('Distribution of SalePrice')
plt.show()
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['SalePrice'])
plt.title('Box Plot of SalePrice')
plt.show()
print(df['SalePrice'].describe())

# Correlation Analysis
correlation_matrix = numerical_features.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
print(correlation_matrix['SalePrice'].sort_values(ascending=False))
```

## Data cleaning

### Subtask:
Clean the data by handling missing values and removing duplicates.


**Reasoning**:
Impute missing numerical values with the median and missing categorical values with "Missing", then remove duplicate rows, excluding the 'Id' column.



```python
# Impute missing numerical values with the median
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].median())

# Impute missing categorical values with "Missing"
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna("Missing")

# Remove duplicate rows, excluding 'Id'
df = df.drop_duplicates(subset=df.columns.difference(['Id']), keep='first')
```

## Data wrangling

### Subtask:
Convert categorical features to numerical representations and investigate potential feature interactions.


**Reasoning**:
Identify categorical columns, apply one-hot encoding to those with fewer unique categories, and label encoding to those with more. Then, create interaction features based on domain knowledge and store them back into the DataFrame.



```python
import pandas as pd

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Apply one-hot encoding or label encoding
for col in categorical_cols:
    if df[col].nunique() < 10:
        dummies = pd.get_dummies(df[col], prefix=col, drop_first=True)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(col, axis=1)
    else:
        # Use label encoding for categorical features with many unique values
        from sklearn.preprocessing import LabelEncoder
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])

# Feature Engineering: Create interaction features
# Example 1: OverallQual * GrLivArea
df['OverallQual_GrLivArea'] = df['OverallQual'] * df['GrLivArea']

# Example 2: OverallQual * GarageCars
df['OverallQual_GarageCars'] = df['OverallQual'] * df['GarageCars']

# Example 3: TotalBsmtSF * 1stFlrSF
df['TotalBsmtSF_1stFlrSF'] = df['TotalBsmtSF'] * df['1stFlrSF']

display(df.head())
```

## Model training

### Subtask:
Train regression models on the training data.


**Reasoning**:
Train the four regression models.



```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize models
linear_model = LinearRegression()
tree_model = DecisionTreeRegressor(random_state=42)
forest_model = RandomForestRegressor(random_state=42)
gb_model = GradientBoostingRegressor(random_state=42)

# Train models
linear_model.fit(X_train, y_train)
tree_model.fit(X_train, y_train)
forest_model.fit(X_train, y_train)
gb_model.fit(X_train, y_train)
```

## Model evaluation

### Subtask:
Evaluate the performance of the trained regression models.


**Reasoning**:
Predict house prices using the trained models and evaluate their performance using MSE and R2 score. Then analyze residuals with residual plots.



```python
import matplotlib.pyplot as plt

# Predict house prices on the test data
linear_pred = linear_model.predict(X_test)
tree_pred = tree_model.predict(X_test)
forest_pred = forest_model.predict(X_test)
gb_pred = gb_model.predict(X_test)

# Calculate MSE and R2 for each model
linear_mse = mean_squared_error(y_test, linear_pred)
linear_r2 = r2_score(y_test, linear_pred)
print(f'Linear Regression - MSE: {linear_mse}, R2: {linear_r2}')

tree_mse = mean_squared_error(y_test, tree_pred)
tree_r2 = r2_score(y_test, tree_pred)
print(f'Decision Tree - MSE: {tree_mse}, R2: {tree_r2}')

forest_mse = mean_squared_error(y_test, forest_pred)
forest_r2 = r2_score(y_test, forest_pred)
print(f'Random Forest - MSE: {forest_mse}, R2: {forest_r2}')

gb_mse = mean_squared_error(y_test, gb_pred)
gb_r2 = r2_score(y_test, gb_pred)
print(f'Gradient Boosting - MSE: {gb_mse}, R2: {gb_r2}')


# Analyze residuals
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
residuals_linear = y_test - linear_pred
plt.scatter(linear_pred, residuals_linear, color='blue', label='Linear Regression')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot for Linear Regression')

plt.subplot(2, 2, 2)
residuals_tree = y_test - tree_pred
plt.scatter(tree_pred, residuals_tree, color='green', label='Decision Tree')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot for Decision Tree')

plt.subplot(2, 2, 3)
residuals_forest = y_test - forest_pred
plt.scatter(forest_pred, residuals_forest, color='orange', label='Random Forest')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot for Random Forest')

plt.subplot(2, 2, 4)
residuals_gb = y_test - gb_pred
plt.scatter(gb_pred, residuals_gb, color='purple', label='Gradient Boosting')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot for Gradient Boosting')

plt.tight_layout()
plt.show()
```

## Model optimization

### Subtask:
Optimize the Gradient Boosting Regressor model.


**Reasoning**:
Optimize the Gradient Boosting Regressor model using GridSearchCV.



```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor

# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10]
}

# Initialize GradientBoostingRegressor
gb_model = GradientBoostingRegressor(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(
    gb_model, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1
)

# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Print the best hyperparameters and best score
print(f"Best hyperparameters: {grid_search.best_params_}")
print(f"Best score (negative MSE): {grid_search.best_score_}")

# Train a new model with the best hyperparameters
best_gb_model = GradientBoostingRegressor(**grid_search.best_params_, random_state=42)
best_gb_model.fit(X_train, y_train)
```
